# Task 4: Most Preferred Investment Avenue

## Objective

Identify the most preferred investment avenue based on participant rankings.

### Method

Participants ranked each investment avenue from 1 (Most Preferred) to 7 (Least Preferred).

The investment avenue with the lowest average ranking is considered the most preferred.

```python
investment_columns = [
    'Mutual_Funds',
    'Equity_Market',
    'Debentures',
    'Government_Bonds',
    'Fixed_Deposits',
    'PPF',
    'Gold'
]

average_rank = df[investment_columns].mean().sort_values()

average_rank
```

```python
investment_df = average_rank.reset_index()

investment_df.columns = ['Investment Avenue', 'Average Rank']

investment_df
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=investment_df,
    x='Investment Avenue',
    y='Average Rank'
)

plt.title("Average Ranking of Investment Avenues")
plt.xlabel("Investment Avenue")
plt.ylabel("Average Rank (Lower = Better)")

plt.xticks(rotation=20)

plt.show()
```

```python
plt.figure(figsize=(9,5))

sns.barplot(
    data=investment_df,
    y='Investment Avenue',
    x='Average Rank'
)

plt.title("Investment Avenue Preference Ranking")

plt.xlabel("Average Rank")

plt.show()
```

```python
best = investment_df.iloc[0]

print("Most Preferred Investment Avenue")
print("--------------------------------")
print(f"Avenue       : {best['Investment Avenue']}")
print(f"Average Rank : {best['Average Rank']:.2f}")
```

## Business Insights

- Investment avenues were ranked by participants from 1 (most preferred) to 7 (least preferred).
- The investment avenue with the lowest average rank represents the highest overall preference.
- The visualization clearly highlights investor priorities and can help financial institutions understand customer preferences.
- Such analyses support better product design and targeted financial planning strategies.

# Task 8: Investment Duration

## Objective

Analyze the investment duration of participants and estimate the average investment duration.

### Method

Since the investment duration is stored as categories, each category is assigned an approximate numeric value:

- Less than 1 year → 0.5 years
- 1–3 years → 2 years
- 3–5 years → 4 years
- More than 5 years → 6 years

These values are used only to estimate the average investment duration.


```python
duration_count = df['Duration'].value_counts()

duration_count
```

```python
duration_percentage = (
    df['Duration']
    .value_counts(normalize=True) * 100
).round(2)

duration_summary = pd.DataFrame({
    'Count': duration_count,
    'Percentage (%)': duration_percentage
})

duration_summary
```

```python
duration_mapping = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6
}

df['Duration_Years'] = df['Duration'].map(duration_mapping)

average_duration = df['Duration_Years'].mean()

print(f"Estimated Average Investment Duration: {average_duration:.2f} years")
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Duration',
    order=duration_count.index
)

plt.title("Investment Duration of Participants")
plt.xlabel("Investment Duration")
plt.ylabel("Number of Participants")

plt.xticks(rotation=15)

plt.show()
```

```python
plt.figure(figsize=(7,7))

plt.pie(
    duration_count,
    labels=duration_count.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Distribution of Investment Duration")

plt.show()
```

## Observation

- Most participants prefer medium-term investments.
- The majority invest for either **1–3 years** or **3–5 years**.
- Very few participants invest for less than one year or more than five years.

## Business Insight

The investment duration pattern indicates that participants generally prefer medium-term investment horizons. Financial institutions can focus on products such as mutual funds, fixed deposits, and balanced portfolios that align with investment periods between one and five years.

## Conclusion

The estimated average investment duration provides an overview of investor commitment. Most participants are medium-term investors, suggesting a preference for balanced growth with moderate risk.

# Task 10: Correlation Analysis

## Objective

Explore the relationships between numerical variables in the dataset using correlation analysis.

### Steps

- Select numerical columns.
- Calculate the correlation matrix.
- Visualize correlations using a heatmap.
- Identify strong positive and negative relationships.
- Summarize key findings.

```python
# Select numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64'])

numerical_columns.head()
```

```python
correlation_matrix = numerical_columns.corr()

correlation_matrix
```

```python
plt.figure(figsize=(12,10))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    linewidths=0.5
)

plt.title("Correlation Matrix of Numerical Variables")

plt.show()
```

```python
corr_pairs = correlation_matrix.unstack()

corr_pairs = corr_pairs[
    corr_pairs != 1
]

corr_pairs = corr_pairs.sort_values(
    ascending=False
)

corr_pairs.head(10)
```

```python
age_corr = correlation_matrix['age'].sort_values(
    ascending=False
)

age_corr
```
